```python
# importing required modules
import boto3

# defining s3
s3 = boto3.resource('s3')

# setting region to the nearest
region = "eu-west-1"

# initialized s3 resource
s3_client = boto3.client('s3', region_name=region)

# downloading the file and saving it in preferred destination
s3.Bucket('blossom-data-engs').download_file('free-7-million-company-dataset.zip', '/home/eddie/Desktop/Blossom_Eddie/Company_Sorted.zip')
```

```python
# importing required modules 
from zipfile import ZipFile
  
# specifying the zip file name 
file_name = "Company_Sorted.zip"
  
# opening the zip file in READ mode
with ZipFile(file_name, 'r') as zip:
  
    # extracting all the files 
    zip.extractall()
```

```python
# importing required modules 
import pandas as pd

# reading file  as csv
companies_sorted = pd.read_csv('companies_sorted.csv', dtype={'a': str})

# summing up the null values in each colmuns
companies_sorted.isnull().sum()

#replacing null values with value 0
companies_sorted.fillna(0, inplace=True)

# dropping rows where there are no domains and assigning it to a new variable
companies_no_domain = companies_sorted.drop(index=companies_sorted[companies_sorted['domain'] == 0].index)

companies_no_domain
```

```python
# exporting dataframe to .csv format
companies_no_domain.to_csv('free-7-million-company-dataset.csv')


# exporting dataframe to .json format with gzip compression
companies_no_domain.to_json('free-7-million-company-dataset-json.gzip', compression = 'gzip')


# converting columns data types to one data type
cols = ["name", "domain","year founded", "industry", "size range", "locality", 
        "country", "linkedin url", "current employee estimate"]

companies_no_domain[cols] = companies_no_domain[cols].apply(pd.to_numeric, errors='coerce', axis=1)

# exporting dataframe to .parquet
companies_no_domain.to_parquet('free-7-million-company-dataset.parquet')
```

```python
# location constraint for s3 bucket
location = {'LocationConstraint': region}

# creating bucket
bucket = "blossom-data-eng-edwin"
s3_client.create_bucket(Bucket=bucket,
                        CreateBucketConfiguration=location)
```

```python
# uploading files to bucket
s3_client.upload_file("free-7-million-company-dataset.csv", bucket, 
                      "project2/free-7-million-company-dataset.csv")

s3_client.upload_file("free-7-million-company-dataset.parquet", bucket, 
                      "project2/free-7-million-company-dataset.parquet")

s3_client.upload_file("free-7-million-company-dataset-json.gzip", bucket, 
                      "project2/free-7-million-company-dataset-json.gzip")
```
